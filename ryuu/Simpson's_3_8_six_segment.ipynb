```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

def integrand(t):
    return 200 * np.log(140000 / (140000 - 2100 * t)) - 9.8 * t

def simpsons_3_8_rule(fun, a, b):
    a, b = min(a, b), max(a, b)
    h = (b - a) / 3
    return (3 * h / 8) * (fun(a) + 3 * fun(a + h) + 3 * fun(a + 2 * h) + fun(b))

a = 8
b = 30
midpoint = (a + b) / 2
total = simpsons_3_8_rule(integrand, a, midpoint) + simpsons_3_8_rule(integrand, midpoint, b)
print(f"Estimated vertical distance: {total:.4f} meters")

x_vals = np.linspace(a, b, 100)
y_vals = integrand(x_vals)

plt.plot(x_vals, y_vals, label="Integrand Curve", color='b')
plt.title("Integrand Curve for Vertical Distance")
plt.xlabel("Time (t) in seconds")
plt.ylabel("Integrand Value")
plt.grid(True)
plt.legend()
plt.show()


x_points = np.linspace(a, b, 6)
y_points = integrand(x_points)
df = pd.DataFrame({'Time (t)': x_points, 'Function Value': y_points})
print(df)
```
